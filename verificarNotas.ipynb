```python
import pandas as pd
from datetime import datetime

# Função para verificar se existem duplicatas de notas na planilha
def verificar_duplicados(planilha):
    # Cria um DataFrame para armazenar as duplicatas
    duplicatas = pd.DataFrame(columns=['N Nota:', 'Dt Venc:'])
    
    # Itera sobre todas as notas para encontrar duplicatas
    for _, row in planilha.iterrows():
        numero_nota = row['N Nota:']
        data_vencimento = row['Dt Venc:']
        
        # Filtra a planilha para ver se existe alguma linha igual à atual (mesmo número e data)
        duplicata = planilha[(planilha['N Nota:'] == numero_nota) & 
                             (planilha['Dt Venc:'] == data_vencimento)]
        
        # Se a duplicata for encontrada e for mais de uma linha, adiciona à lista de duplicatas
        if len(duplicata) > 1 and duplicata.iloc[0].name != row.name:
            duplicatas = pd.concat([duplicatas, duplicata])
    
    # Remove linhas duplicadas para não mostrar o mesmo valor várias vezes
    duplicatas = duplicatas.drop_duplicates()
    
    # Se existirem duplicatas, retorna elas; caso contrário, retorna uma mensagem de nenhuma duplicata
    if not duplicatas.empty:
        return duplicatas
    else:
        return "Não há notas duplicadas."

# Carregar a planilha existente (se já existir)
caminho_arquivo = 'teste.xlsx'

# Tenta carregar a planilha; caso contrário, exibe mensagem de erro
try:
    planilha = pd.read_excel(caminho_arquivo)
except FileNotFoundError:
    print("Arquivo não encontrado. Por favor, verifique o caminho do arquivo.")

# Verificar e mostrar as duplicatas
duplicatas = verificar_duplicados(planilha)

if isinstance(duplicatas, pd.DataFrame):
    print("Notas Duplicadas Encontradas:")
    display(duplicatas)
else:
    display(duplicatas)
```
